## Summary Statistics

```python
# Generate a summary statistics table of mean, median, variance, standard deviation, and SEM of the tumor volume for each regimen
# This method is the most straighforward, creating multiple series and putting them all together at the end.
#extract mean for the mean of each regimen
#THIS IS THE 1ST METHOD

groupby_regimen = clean_df.groupby("Drug Regimen")

#calculate the statistics
mean_group = groupby_regimen["Tumor Volume (mm3)"].mean()
med_group = groupby_regimen["Tumor Volume (mm3)"].median()
var_group = groupby_regimen["Tumor Volume (mm3)"].var()
std_group = groupby_regimen["Tumor Volume (mm3)"].std()
sem_group = groupby_regimen["Tumor Volume (mm3)"].sem()


#collate all the data into dataframe

summary_statistics2 = pd.DataFrame({"Mean of Tumor Volume":mean_group,
                                   "Median of Tumor Volume" : med_group,
                                   "Variance of Tumor Volume": var_group,
                                    "STD of Tumor Volume" : std_group,
                                      "SEM of Tumor Volume" : sem_group})

summary_statistics2
```

```python
# Generate a summary statistics table of mean, median, variance, standard deviation, and SEM of the tumor volume for each regimen
# This method produces everything in a single groupby function
#use AGGREGATION METHOD
mean_tumor = clean_df.groupby("Drug Regimen").mean()["Tumor Volume (mm3)"]

#extract the median for each regimen
median_tumor = clean_df.groupby("Drug Regimen").median()["Tumor Volume (mm3)"]

#extract the variance for each regimen
variance_tumor = clean_df.groupby("Drug Regimen").var()["Tumor Volume (mm3)"]

#extract the standard deviation for each regimen
std_tumor = clean_df.groupby("Drug Regimen").std()["Tumor Volume (mm3)"]

#extract the SEM for each regimen
sem_tumor = clean_df.groupby("Drug Regimen").sem()["Tumor Volume (mm3)"]

summary_statistics1 = pd.DataFrame({"Mean of Tumor Volume":mean_tumor,
                                   "Median of Tumor Volume" : median_tumor,
                                   "Variance of Tumor Volume": variance_tumor,
                                    "STD of Tumor Volume" : std_tumor,
                                      "SEM of Tumor Volume" : sem_tumor})


summary_statistics1 
```

```python
# Calculate the final tumor volume of each mouse across four of the treatment regimens:  
# Capomulin, Ramicane, Infubinol, and Ceftamin
# Start by getting the last (greatest) timepoint for each mouse 
# Merge this group df with the original dataframe to get the tumor volume at the last timepoint
```

```python
#using the dataframe with the cleaned data
clean_df
```

```python
#Select the data that is under 4 given regimen only : Capomulin, Ramicane, Infubinol, and Ceftamin
four_treatment_df = clean_df.loc[(clean_df["Drug Regimen"]=="Capomulin")|(clean_df["Drug Regimen"]=="Ramicane")|(clean_df["Drug Regimen"]=="Infubinol") | (clean_df["Drug Regimen"]=="Ceftamin")]
four_treatment_df
```

```python
#using groupby function and sorted the data based on MOUSE ID, and their TIMEPOINT -- diplay the last TIMEPOINT as the first value
four_treatment_df2 = four_treatment_df.sort_values(["Mouse ID","Drug Regimen","Timepoint"], ascending = False)
four_treatment_df2
```

```python
#from the sorted dataframe, remove the duplicates (assuming this will keep the first row)
four_treatment_df3 = four_treatment_df2.drop_duplicates(subset = ["Mouse ID"])
four_treatment_df3                                  
```

```python
#checking if the drop duplicate functions is done correctly, same no. of mice as the unique count = correct
four_treatment_df3.shape
```

```python
four_treatment_df3 = four_treatment_df3.reset_index(drop=True)
four_treatment_df3
```

```python
# Put treatments into a list for for loop (and later for plot labels)
list_of_regimen = ["Capomulin", "Ramicane", "Infubinol", "Ceftamin"]

# Create empty list to fill with tumor vol data (for plotting)
empty_list  = []

# Calculate the IQR and quantitatively determine if there are any potential outliers. 
for value in list_of_regimen:
    
    
    # Locate the rows which contain mice on each drug and get the tumor volumes
    select_treatment = four_treatment_df3.loc[four_treatment_df3["Drug Regimen"] == value, "Tumor Volume (mm3)"]
    
    # add subset 
    empty_list.append(select_treatment)

    # Determine outliers using upper and lower bounds
    quartiles = select_treatment.quantile([.25,.5,.75])
    lowerq = quartiles[0.25]
    upperq = quartiles[0.75]
    iqr = upperq - lowerq
    lower_bound = lowerq - (1.5*iqr)
    upper_bound = upperq + (1.5*iqr)

    print(f'The lower quartile of the final tumor volume under Drug Regimen {value} is {lowerq}')
    print(f'The upper quartile of the final tumor volume under Drug Regimen {value} is {upperq}')
    print(f'The interquartile range of Drug Regimen {value} is {iqr}')
    print(f'The median range of Drug Regimen {value} is {quartiles[0.5]}')
    print(f'Value below {lower_bound} for Drug Regimen {value} could be outliers')
    print(f'Value above {upper_bound} Drug Regimen {value} could be outliers')
```

## Correlation and Regression

```python
# Calculate the correlation coefficient and linear regression model 
# for mouse weight and average tumor volume for the Capomulin regimen

weightg = mice_weight_df["Weight (g)"]
tumorvolume = mice_weight_df["Tumor Volume (mm3)"]
correlation = st.pearsonr(weightg, tumorvolume)



(slope, intercept, rvalue, pvalue, stderr) = linregress (weightg, tumorvolume)
regress_values = (weightg * slope) +intercept
line_eq = "y =  "+ str(round(slope, 2)) + "x +  "+ str(round(intercept,2))
plt.scatter(weightg, tumorvolume)
plt.plot (weightg, regress_values, "r-")
plt.annotate(line_eq, (8,8),fontsize = 50, color = "red")
plt.xlabel("Mice Weight (g)")
plt.ylabel("Tumor Volume (mm3)")
plt.show()



# The correlation coefficient for the study
print( f'The correlation between both factors is {round(correlation[0],2)}')
```
